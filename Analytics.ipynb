```python
import pandas as pd
df = pd.read_csv('shein_mens_fashion.csv')
df.head()
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt
rating_ranges = [(0, 1), (1, 2), (2, 3), (3, 4), (4, 5)]
reviews_count_by_range = [((df['average_rating'] >= low) & (df['average_rating'] < high)).sum() for low, high in rating_ranges]
labels = [f'{low}-{high}' for low, high in rating_ranges]

plt.figure(figsize=(10, 6))
plt.bar(labels, reviews_count_by_range, color='skyblue')
plt.xlabel('Average Rating Range')
plt.ylabel('Reviews Count')
plt.title('Distribution of Reviews Count by Average Rating Range')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
